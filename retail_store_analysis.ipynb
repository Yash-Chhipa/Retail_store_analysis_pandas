```python
import pandas as pd
df = pd.read_csv('train.csv')
```

1. Find the total sales for each city 

```python

city_sales = df.groupby("City")["Sales"].sum()
print(city_sales)
```

2.Identify the top 5 products with highest sales.

```python

top_products = (
    df.groupby("Product Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

print(top_products)
```

3.calculate total sales for each category

```python

total_sales_category = df.groupby("Category")["Sales"].sum()
print(total_sales_category)
```

4.Find the customer who generated the highest total sales

```python

top_customer = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(1)

print(top_customer)
```

5.Determine the city with highest revenue.

```python
top_city = df.groupby("City")["Sales"].sum().sort_values(ascending=False).head(1)

print(top_city)
```

6.Find the total sales for each subcategory 

```python
subcat_sales = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)
print(subcat_sales)
```

7. calculate the monthly sales trend for each subcategory 

```python
# Ensure date column is datetime
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")

# Monthly sales trend for each sub-category
monthly_trend = df.pivot_table(
    values="Sales",
    index=df["Order Date"].dt.to_period("M"),
    columns="Sub-Category",
    aggfunc="sum"
)

print(monthly_trend)
```

8. Identify orders with sales greater than 1000

```python
high_value_orders = df[df["Sales"] > 1000]

print(high_value_orders)
```

9.calculate the average sales per customer

```python
avg_sales_per_customer = df.groupby("Customer Name")["Sales"].mean()

print(avg_sales_per_customer)
```

10. Identify the least-selling product(based on total sales)

```python
least_product = df.groupby("Product Name")["Sales"].sum().sort_values().head(1)

print(least_product)
```
