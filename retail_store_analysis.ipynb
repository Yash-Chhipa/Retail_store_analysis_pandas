```python
import pandas as pd
df = pd.read_csv('train.csv')
```

```python
# total sales for each city
city_sales = df.groupby("City")["Sales"].sum()
print(city_sales)
```

```python
#Identify the top 5 products with highest sales.
top_products = (
    df.groupby("Product Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

print(top_products)
```

```python
#Calculate the average discount given by category.
df.groupby("Category")["Discount"].mean()
```

```python
#Find the customer who purchased the most products.
top_customer = (
    df.groupby("Customer Name")
    .size()
    .sort_values(ascending=False)
    .head(1)
)

print(top_customer)
```

```python
#Determine the city with highest revenue.
top_city = (
    df.groupby("City")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(1)
)

print(top_city)
```

```python
#Find the total quantity sold per category.
category_sales = df.groupby("Category").size()
print(category_sales)
```

```python
#Calculate monthly sales trend using order_date.monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()
# Convert Order Date to datetime
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")

# Monthly sales trend
monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()

print(monthly_sales)
```

```python
#Identify orders where discount > 20%.
df[df["Discount"] > 0.2]
```

```python
#Find the average order value per customer.
order_values = df.groupby(["Order ID","Customer Name"])["Sales"].sum().reset_index()
avg_order_value = order_values.groupby("Customer Name")["Sales"].mean()
print(avg_order_value)
```

```python
#Identify the least selling product.
least_product = df.groupby("Product Name")["Sales"].sum().sort_values().head(1)
print(least_product)
```
